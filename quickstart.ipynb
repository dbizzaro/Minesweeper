```python
def ask_mistral(prompt):
  chat_response = client.chat.complete(
    model=model,
    messages=[{"role":"user", "content":prompt}]
  )
  return chat_response.choices[0].message.content
```

```python
topics1 = ['nature', 'traveling', 'death', 'love', 'family', 'art', 'life', 'desire']
topics2 = ['Brazil', 'an ostrich', 'my friend Giorgio', 'a bot', 'Alexandre Dumas', 'the larinx', 'illuminism', 'an old man sitting on a bench']

#quatrains = ['a quatrain', 'two quatrains', 'three quatrains', 'four quatrains']
#tercets = ['a tercet', 'two tercets', 'three tercets', 'four tercets']

stanzas = ['three quatrains', 'four tercets']

meters = ['iambic pentameters', 'trochaic tetrameters', 'anapestic trimeter'] #'spondaic dimeter', 'amphibrachic tetrameter', 'choriambic pentameter', 'iambic hexameter (alexandrine)', 'trochaic trimeter', 'anapestic tetrameter', 'dactylic pentameter']

rhyme_schemes_quatrain = ['ABBA CDDC EFFE', 'ABAB CDCD EFEF', 'ABBA CDCD EEFF', 'ABAA CDCC EFEE'] # 'ABAB BABA AABB', 'ABCD ABCD ABCD']
rhyme_schemes_tercet = ['ABA CDC EFE GHG', 'ABA BCB CDC DED', 'AAA BBB CCC DDD', 'ABC ABC ABC ABC'] #, 'AAB BAA CCD DCC', 'ABA BAB CDC DCD']

data = []
for stanza in stanzas:
  rhyme_scheme = rhyme_schemes_quatrain if stanza == 'three quatrains' else rhyme_schemes_tercet
  for meter in meters:
    for scheme in rhyme_scheme:
      for topic in topics1:
        prompt = f"Write a poem about {topic}. It should be made of {stanza} of {meter}. The rhyme scheme should be {scheme}. Generate only the poem, and do not indicate the rhyme scheme."
        poem = ask_mistral(prompt)
        print(prompt)
        print(poem)
        print()
        data.append([stanza, meter, scheme, topic, prompt, poem])
        time.sleep(5)
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=['Stanza', 'Meter', 'Scheme', 'Topic', 'Prompt', 'Poem'])
df.to_csv('mistral_poems.csv', index=False)
```
